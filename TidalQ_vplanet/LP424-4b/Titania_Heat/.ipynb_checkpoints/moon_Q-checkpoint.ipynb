```python
import numpy as np
from ism_abs import mm_abs
```

```python
E = np.linspace(0.03, 10.0, 1000)
```

```python
plt.plot(E, mm_abs(0.5E22, E), color='blue', lw=1.5, label=r'$\rm{0.5}$');
plt.plot(E, mm_abs(3.0E22, E), color='red', lw=1.5, label=r'$\rm{3.0}$');

#for i, el in enumerate(E):
#    plt.scatter(el, mm_abs(0.5E22, el), color='blue')
#    plt.scatter(el, mm_abs(3.0E22, el), color='red')


plt.xlabel(r'$\rm{Energy \, [keV]}$', fontsize=17)
plt.ylabel(r'$\rm{e^{-N_{H} \ \sigma(E)}}$', fontsize=22)

plt.xlim(0.05, 3.0);
plt.ylim(1.E-11, 1.E0);

plt.xscale('log');
plt.yscale('log');

plt.xticks([0.1, 1.0], ['0.1', '1'], fontsize=13);
plt.yticks(fontsize=13);
plt.yticks([1.E-10, 1.E-8, 1.E-6, 1.E-4, 1.E-2, 1.E0])

plt.tick_params(which='both', width=1.5)
plt.tick_params(which='major', length=10)
plt.tick_params(which='minor', length=5)


leg_title = r'$\rm{N_{H} \, [10^{22} \ cm^{2}]}$'
l = plt.legend(loc=2, title = leg_title, frameon=False, bbox_to_anchor=(0.02, 0.98));
setp(l.get_texts(), fontsize=15); 
setp(l.get_title(), fontsize=15);

plt.tight_layout()
plt.savefig('ism_abs.pdf', bbox_to_inches = 'tight')
```
